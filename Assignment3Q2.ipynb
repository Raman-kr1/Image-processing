```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1=cv.imread('/content/drive/MyDrive/Assignment_3_images/cctv1.JPG')
img2=cv.imread('/content/drive/MyDrive/Assignment_3_images/cctv2.JPG')
img3=cv.imread('/content/drive/MyDrive/Assignment_3_images/cctv3.JPG')
img4=cv.imread('/content/drive/MyDrive/Assignment_3_images/cctv4.JPG')
```

```python
number= input()
```

```python
if(number=='1') :
  img=img1
elif (number=='2') :
  img=img2
elif (number=='3') :
  img=img3
elif (number=='4') :
  img=img4
else : print('Enter a Valid Number')
```

```python
cv2_imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
filtered = cv.GaussianBlur(gray, (5, 5), 0)
cv2_imshow(filtered)
```

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
enhanced_image = cv.equalizeHist(gray_image)
cv2_imshow(enhanced_image)
```

```python
filtered_image = cv.medianBlur(enhanced_image, 5)
edges = cv.Canny(filtered_image, 100, 200)
cv2_imshow(edges)
```
