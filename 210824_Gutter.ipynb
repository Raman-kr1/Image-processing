```python
import cv2 as cv
import numpy as np

from google.colab.patches import cv2_imshow
```

```python
# Taking image input

img_number = 3

img = cv.imread(f'gutters{img_number}.JPG')
cv2_imshow(img)
```

```python
# Converting the color image to grayscale

gray = cv.cvtColor(img, code = cv.COLOR_BGR2GRAY)
```

```python
# Using adaptive threshold method to remove the gutter

if(img_number == 1):
  kernel_size = 13
  constant = 11
  final = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, kernel_size, constant)
```

```python
# Using adaptive threshold method to remove the gutter

if(img_number == 2):
  kernel_size = 11
  constant = 15
  final = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, kernel_size, constant)
```

```python
# Using adaptive threshold method to remove the gutter

if(img_number == 3):
  kernel_size = 17
  constant = 19
  final = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, kernel_size, constant)
```

```python
cv2_imshow(final)
cv.imwrite(f'./cleaned_gutters{img_number}.JPG', final)
```
